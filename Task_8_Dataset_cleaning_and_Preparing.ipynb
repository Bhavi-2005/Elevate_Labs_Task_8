```python
import pandas as pd
```

```python
df = pd.read_csv("/content/sample_-_superstore.csv", encoding='latin-1')
```

```python
df.head()
df.info()
df.describe()
```

```python
df.dropna(subset=["Order Date", "Region", "Category", "Sales", "Profit"], inplace=True)
```

```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df['Month-Year'] = df['Order Date'].dt.strftime('%b-%Y')
df['Month-Sort'] = df['Order Date'].dt.to_period('M')
```

```python
df.to_csv("Superstore_Sales_Cleaned.csv", index=False)
```
